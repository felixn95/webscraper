```python
import warnings
import pandas as pd
from data_analysis.DataProcessor import DataProcessor

warnings.simplefilter(action='ignore', category=FutureWarning)

# read in combined stock data
preprocessor = DataProcessor()
# create df to current date
df = preprocessor.process_nrw_data("06_22")
# remove bike sale since it is not relevant
df = preprocessor.remove_fahrrad_sale(df)
# Ensure 'price' is numeric
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import os

# Ensure 'price' is numeric
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Confirm 'timestamp' is datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Identify stock events
def identify_stock_events(df):
    # Sort the dataframe by 'id', 'store_skuId', 'store_storeId', and 'timestamp'
    df = df.sort_values(by=['id', 'store_skuId', 'store_storeId', 'timestamp'])
    
    # Identify stockout events
    df['stockout'] = (
        (df['store_quantity'] == 0) & 
        (df['store_quantity'].shift(1) > 0) & 
        (df['store_skuId'].shift(1) == df['store_skuId']) & 
        (df['store_storeId'].shift(1) == df['store_storeId'])
    )
    
    # Identify restock events
    df['restock'] = (
        (df['store_quantity'] > 0) & 
        (df['store_quantity'].shift(1) == 0) & 
        (df['store_skuId'].shift(1) == df['store_skuId']) & 
        (df['store_storeId'].shift(1) == df['store_storeId'])
    )
    
    return df

# Remove outliers using IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    filter = (df[column] >= Q1 - 1.5 * IQR) & (df[column] <= Q3 + 1.5 * IQR)
    return df.loc[filter]

# Filter by minimum sales events
def filter_by_sales_threshold(df, min_sales_events):
    # Count the number of sales events per product variant (store_skuId and store_storeId combination)
    sales_counts = df.groupby(['store_skuId', 'store_storeId']).size()
    
    # Filter to include only those with at least min_sales_events sales events
    valid_variants = sales_counts[sales_counts >= min_sales_events].index
    
    return df[df.set_index(['store_skuId', 'store_storeId']).index.isin(valid_variants)]

# Calculate restock times with options to remove outliers and filter by minimum sales events
def calculate_restock_times(df, remove_outliers_flag=False, min_sales_events=None):
    restock_times = []
    grouped = df.groupby(['id', 'store_skuId', 'store_storeId'])
    
    for _, group in grouped:
        group = group.sort_values(by='timestamp')
        stockouts = group[group['stockout']].copy()
        restocks = group[group['restock']].copy()
        
        for stockout_idx, stockout_row in stockouts.iterrows():
            restock_after_stockout = restocks[restocks['timestamp'] > stockout_row['timestamp']]
            if not restock_after_stockout.empty:
                restock_time = restock_after_stockout.iloc[0]['timestamp'] - stockout_row['timestamp']
                restock_times.append({
                    'id': stockout_row['id'],
                    'store_skuId': stockout_row['store_skuId'],
                    'store_storeId': stockout_row['store_storeId'],
                    'main_category': stockout_row['main_category'],
                    'sub_category': stockout_row['sub_category'],
                    'brand': stockout_row['brand'],
                    'stockout_time': stockout_row['timestamp'],
                    'restock_time': restock_after_stockout.iloc[0]['timestamp'],
                    'time_to_restock': restock_time.total_seconds() / (60 * 60 * 24)  # convert to days
                })
    
    restock_times_df = pd.DataFrame(restock_times)
    
    # Remove outliers if flag is set
    if remove_outliers_flag:
        restock_times_df = remove_outliers(restock_times_df, 'time_to_restock')
    
    # Filter by minimum sales events if specified
    if min_sales_events is not None:
        restock_times_df = filter_by_sales_threshold(restock_times_df, min_sales_events)
    
    return restock_times_df

# Method to show analysis with options for outliers and minimum sales events
def show_analysis(df, remove_outliers_flag=False, min_sales_events=None):
    # Apply the identify_stock_events function to the dataframe
    df = identify_stock_events(df)
    
    restock_times_df = calculate_restock_times(df, remove_outliers_flag, min_sales_events)
    
    # Average restock time by main and sub category
    avg_restock_time_by_category = restock_times_df.groupby(['main_category', 'sub_category'])['time_to_restock'].mean()
    
    # Create a complete index from the existing combinations in the data
    all_categories = restock_times_df.set_index(['main_category', 'sub_category']).index.drop_duplicates()
    avg_restock_time_by_category = avg_restock_time_by_category.reindex(all_categories).fillna(0)
    
    # Restock times distribution
    plt.figure(figsize=(10, 6))
    restock_times_df['time_to_restock'].hist(bins=50)
    plt.title('Distribution of Restock Times (in days)')
    plt.xlabel('Time to Restock (days)')
    plt.ylabel('Frequency')
    plt.show()
    
    # Benchmarking
    benchmark_by_store_category = restock_times_df.groupby(['store_storeId', 'main_category', 'sub_category'])['time_to_restock'].mean().unstack(level=['main_category', 'sub_category'])
    benchmark_by_store_category = benchmark_by_store_category.reindex(columns=all_categories, fill_value=0)
    
    return restock_times_df, avg_restock_time_by_category, benchmark_by_store_category

# Create the directory if it doesn't exist
output_dir = 'data_restock_only_stockouts'
os.makedirs(output_dir, exist_ok=True)

# Analysis with outliers
print("Analysis with outliers:")
restock_times_df_with_outliers, avg_restock_time_with_outliers, benchmark_with_outliers = show_analysis(df, remove_outliers_flag=False)
print("Average Restock Time by Main and Sub Category (in days):")
display(avg_restock_time_with_outliers)
print("Benchmark by Store, Main Category, and Sub Category (in days):")
display(benchmark_with_outliers)

# Analysis without outliers
print("Analysis without outliers:")
restock_times_df_without_outliers, avg_restock_time_without_outliers, benchmark_without_outliers = show_analysis(df, remove_outliers_flag=True)
print("Average Restock Time by Main and Sub Category (in days):")
display(avg_restock_time_without_outliers)
print("Benchmark by Store, Main Category, and Sub Category (in days):")
display(benchmark_without_outliers)

# Analysis with outliers and minimum sales threshold
print("Analysis with outliers and minimum sales threshold:")
restock_times_df_with_outliers_min_sales, avg_restock_time_with_outliers_min_sales, benchmark_with_outliers_min_sales = show_analysis(df, remove_outliers_flag=False, min_sales_events=2)
print("Average Restock Time by Main and Sub Category with Minimum Sales (in days):")
display(avg_restock_time_with_outliers_min_sales)
print("Benchmark by Store, Main Category, and Sub Category with Minimum Sales (in days):")
display(benchmark_with_outliers_min_sales)

# Analysis without outliers and minimum sales threshold
print("Analysis without outliers and minimum sales threshold:")
restock_times_df_without_outliers_min_sales, avg_restock_time_without_outliers_min_sales, benchmark_without_outliers_min_sales = show_analysis(df, remove_outliers_flag=True, min_sales_events=2)
print("Average Restock Time by Main and Sub Category with Minimum Sales (in days):")
display(avg_restock_time_without_outliers_min_sales)
print("Benchmark by Store, Main Category, and Sub Category with Minimum Sales (in days):")
display(benchmark_without_outliers_min_sales)
```
